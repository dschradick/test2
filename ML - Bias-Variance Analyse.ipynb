```python
import matplotlib.pyplot as plt
import scikitplot as skplt
from sklearn.datasets import load_breast_cancer as load_data

X, y = load_data(return_X_y=True)
```

### Modelle

```python
### Drei verschiedene Modell-Klassen mit verschiedenen Eigenschaften

# Hoher Bias, geringe Varianz
from sklearn.linear_model import LinearRegression

# Geringer Bias, hohe Varianz
from sklearn.tree import DecisionTreeRegressor

# Geringer Bias, geringe Varianz
from sklearn.ensemble import RandomForestClassifier
```

### Linear Regression

```python
# Hoher Bias
linear = LinearRegression()
skplt.estimators.plot_learning_curve(linear, X, y);

plt.ylim((0.4,1.1));
```

### Entscheidungsbaum

```python
# Baum mit nierdrigerer Varianz

rf = DecisionTreeRegressor(max_depth=2)    
skplt.estimators.plot_learning_curve(rf, X, y)

plt.ylim((0.4,1.1))
```

```python
# Baum mit hoher Varianz

rf = DecisionTreeRegressor(max_depth=10)   
skplt.estimators.plot_learning_curve(rf, X, y,)

plt.ylim((0.4,1.1))
```

### Random Forest

```python
rf = RandomForestClassifier(n_estimators=2)
skplt.estimators.plot_learning_curve(rf, X, y)
plt.ylim((0.8,1.1))
```

```python
rf = RandomForestClassifier(n_estimators=1000)
skplt.estimators.plot_learning_curve(rf, X, y)
plt.ylim((0.8,1.1));
```
